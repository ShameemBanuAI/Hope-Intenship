# Importing Libraries

```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("Inputdata.csv")
```

```python
df
```

# Checking for null values

```python
df.isnull().sum()
```

```python
# Finding the datatype of the columns
df.info()
```

# Result type alone is in object Datatype having null value

```python
df['Result Type'] = df['Result Type'].fillna('NA')
```

# Filling null values with median without changing info

```python
# Replace NaN values in multiple columns with 1
columns_to_fill = ['Results', 'Link clicks', 'Landing page views']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].median())
```

# Using SimpleImputer to change many null values to 0

```python

imp = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=0)

# Fit and transform the DataFrame, and retain the column names
dataFrame = pd.DataFrame(imp.fit_transform(df), columns=df.columns)

# Display the result
dataFrame.head(10)
```

```python
dataFrame.isnull().sum()
```

```python
colName=["Date","Campaign Name","AD Spent","No.of Leads","Cost Per Lead","CTR","Link Click","LP Views","member passed","LP Conversion %"]
```

```python
# Creating output data frame with specified columns
op_df=pd.DataFrame(columns=colName)
```

```python
op_df
```

# Working for Date column in output dataframe

```python
def convert_date(date_str):
    date_object = datetime.strptime(date_str, '%m/%d/%Y')
    day_of_month = str(int(date_object.strftime('%d')))  # Remove leading zeros
    month_name = date_object.strftime('%B')
    day_of_week = date_object.strftime('%A')
    return f"{day_of_month} {month_name} ({day_of_week})"

op_df['Date'] = dataFrame['Day'].apply(convert_date)

op_df
```

# Working for other columns in the output frame

```python
op_df['Campaign Name']=dataFrame['Campaign name']
op_df['AD Spent']=dataFrame['Amount spent (INR)']
op_df['No.of Leads']=dataFrame['Results']
```

```python
op_df
```

```python
op_df['Cost Per Lead']=op_df['AD Spent']/op_df['No.of Leads']
```

```python
op_df
```

```python
op_df['CTR']=dataFrame['CTR (link click-through rate)']
op_df['Link Click']=dataFrame['Link clicks']
op_df['LP Views']=dataFrame['Landing page views']
```

```python
op_df
```

```python
op_df['member passed']=(op_df['LP Views']/op_df['Link Click'])*100
```

```python
op_df['LP Conversion %']= (op_df['No.of Leads']/op_df['LP Views'])*100
```
